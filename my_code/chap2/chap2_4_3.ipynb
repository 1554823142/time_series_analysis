## 数据平滑

```python
import pandas as pd

air = pd.read_csv('../../demo_code/Ch02/data/AirPassengers.csv', names=['Date', 'Passengers'])
# 自定义列名，可以在读取文件时使用 header=0 参数来告诉 Pandas 读取文件的第一行作为数据，并通过 names 参数覆盖列名
air
```

```python
# 使用衰减因子进行平滑处理 alpha为平滑因子 alpha越大，平滑结果就越接近当前值
air['Smooth.5'] = air['Passengers'].ewm(alpha=0.5).mean()
air['Smooth.9'] = air['Passengers'].ewm(alpha=0.9).mean()
air
```
